# Load Uber.csv dataset.

```python

import pandas as pd 
uber=pd.read_csv("C:/Users/mohan reddy/Downloads/Uber_Drives_2016.csv") 
uber.head()
```

# Display rides where CATEGORY == 'Business'.

```python
business_rides = uber[uber['CATEGORY*'] == 'Business']
print(business_rides)
```

#  Show top 5 rides with the longest distance (MILES).


```python
top5_longest_rides = uber.sort_values(by='MILES*', ascending=False).head(5)
print(top5_longest_rides)
```

# Sort trips by distance in descending order.

```python
trips = uber.sort_values(by='MILES*', ascending=False)
trips
```
